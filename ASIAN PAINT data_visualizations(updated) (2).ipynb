```python
#timewise performance 
import seaborn as sns
d=df['timestamp']
sns.relplot(data=df, x='timestamp', y='close', hue='volume', height=10, aspect=4/3)
```

```python
#date wise open price performance
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
df['date'] = pd.to_datetime(df.timestamp)  # d.Date = d.Date.astype('datetime64[ns]')
ax.scatter(x=df['date'], y=df['open'])
plt.show()
```

```python
#date wise high,low performance
start_date = "2019-04-30"
end_date = "2021-04-30"

mask = (df['date'] >= start_date) & (df['date'] <= end_date)
asianpaint_df1 =df.loc[mask]

plt.figure(figsize=(15,10));
plt.plot(asianpaint_df1.date, asianpaint_df1.high);
plt.plot(asianpaint_df1.date,asianpaint_df1.low);
plt.title("Trends of Asian Paint Stock 2020-2021");
plt.xlabel('2019-2021');
plt.ylabel('Price in INR');
plt.legend(['High','Low']);
```

```python
#volume price 2019-2021
asianpaint_df1['volume'].plot()
plt.title("Asian Paint volume Price April-2019 - 2021")
plt.show()
```

```python
#open,close,high,low performance
plt.figure(figsize=(20,7));
asianpaint_df1['open'].plot(label = 'Asian Paint Open Price')
asianpaint_df1['close'].plot(label = 'Asian Paint Close Price')
asianpaint_df1['high'].plot(label= 'Asian Paint High Price')
asianpaint_df1['low'].plot(label= 'Asian Paint Low Price')
plt.legend()
plt.title('Asian Paint Prices')
plt.ylabel("Stock Price")
plt.show()
```

```python
#date wise high 
plt.figure(figsize=(15,12));
plt.bar(asianpaint_df1.date, asianpaint_df1.high);
plt.show()
```

```python
plt.bar(asianpaint_df1.date,asianpaint_df1.volume,width=2.0)
plt.show()
```

```python
#open price for a timeframe(2019-2021)
asianpaint_df1['open'].plot()
plt.title("Asian Paint Open Price April-2019-2021")
plt.show
```

```python
AvgHigh = asianpaint_df1.describe()['high']['mean']
AvgLow = asianpaint_df1.describe()['low']['mean']
AvgVolume = asianpaint_df1.describe()['volume']['mean']

print("Average High Asian Paints Stock: ",AvgHigh,
     "\nAverage Low Asian Paints: ",AvgLow,
     "\nAverage Volume for Asian Paint: ", AvgVolume)
```

```python
#close price performance(2019-2021)
asianpaint_df1['close'].plot()
plt.title("Asian Paint Close Price April-2019 - 2021")
plt.show
```

```python
#top high prices
high_10_price = asianpaint_df1.sort_values(by='high', ascending=False).head(10)
high_10_price
```
